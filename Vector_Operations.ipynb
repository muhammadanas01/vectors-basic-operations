Addition of Two Vectors

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])

sum = vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units ="xy", angles='xy', scale=1, color="b")
plt.quiver(0,0,3,-2, scale_units ="xy", angles='xy', scale=1, color="y")
plt.quiver(0,0,5,1, scale_units ="xy", angles='xy', scale=1, color="r")
plt.xlim(-10,10)
plt.ylim(-10,10)

plt.show()
```

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])

diff = vector_1 - vector_2
print(diff)
```

DOT PRODUCT OF TWO VECTORS

```python
a = np.array([2,3])
b = np.array([4,4])
```

```python
a_dot_b = np.dot(a,b)

print(a_dot_b)
```

```python
c=np.array([40,20,35])

d = np.array([30,43,11])

c_dot_d = np.dot(c,d)

print(c_dot_d)
```

CROSS PRODUCT

```python
a_cross_b = np.cross(a,b)
```

```python
print(a_cross_b)
```

```python
c_cross_d = np.cross(c,d)

print(c_cross_d)
```

Projection Of Vector

```python

a = np.array([2,3])

v = np.array([8,-7])


#Magnitude of v vector
# v_square = sum(v**2)
mag_v = np.sqrt(np.sum(v**2))

proj_of_a_on_v = (np.dot(a,v)/(mag_v**2))*v

print(mag_v)
print(proj_of_a_on_v)
```
